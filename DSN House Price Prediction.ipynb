```python
#Loading and printting the first 5 observations of the training set  
train = pd.read_csv("Desktop/Udacity Nano Degree/Housing_dataset_train.csv")
train.head()
```

```python
#Loading and printting the first 5 observations of the testing set
test = pd.read_csv("Desktop/Udacity Nano Degree/Housing_dataset_test.csv")
test.head()
```

# EXPLOXITORY DATA ANALYSIS OF THE TRAIN DATA

```python
#Checking for missing values in the training data
train.isnull().sum()
```

```python
#Treating the numerical missing values data with the mean of each of their columns
#Treating the categorical missing values data with the mode of each of their columns
mode_loc = train['loc'].mode()
mode_title = train['title'].mode()
mean_bedroom = train["bedroom"].mean()
mean_bathroom = train["bathroom"].mean()
mean_parking_space = train["parking_space"].mean()

#Filling the missing data
train['loc'].fillna(mode_loc, inplace = True)
train['title'].fillna(mode_title, inplace = True)
train['bedroom'].fillna(mean_bedroom, inplace = True )
train['bathroom'].fillna(mean_bathroom, inplace = True )
train['parking_space'].fillna(mean_parking_space, inplace = True )
```

```python
#Checking if their is any other missingness
train.isna().sum()
```

```python
train['loc'].fillna('Kaduna', inplace= True)
train['title'].fillna('Flat', inplace = True)
```

```python
#Checking if their is any other missingness
train.isna().sum()
```

```python
#Treating the categorical data 
categorical_data = train[['loc','title']]
```

```python
#Print the number of unique values of the in each of the categorical feature
categorical_data.nunique()
```

```python
#Count the value of the loc column
categorical_data['loc'].value_counts()
```

```python
#Count the value of the title column
categorical_data['title'].value_counts()
```

```python
#Appling ONE-Hot Encoding to take care of the loc column and assing it to the veriable states
states = train['loc'].str.get_dummies(sep = ' ')
states
```

```python
#Dropping the loc column and concatinating the state with the train data
train = train.drop('loc', axis = 1)
train = pd.concat([train, states], axis = 1)
#One-Hot Encoding to the title column
train = pd.get_dummies(data = train, columns = ['title'])
train
```

```python
print("New Shape of the training data: ",train.shape)
```
